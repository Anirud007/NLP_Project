## Scrape

```python
import pandas as pd
from textblob import TextBlob
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
```

```python
from google_play_scraper import app, reviews, Sort

app_id = 'in.goindigo.android'  

app_details = app(app_id)
print("App Name:", app_details['title'])
print("Description:", app_details['description'])

result, continuation_token = reviews(
    app_id,
    lang='en',  
    country='us',  
    sort=Sort.NEWEST,
    count=2000   
)

reviews_df = pd.DataFrame(result)
```

## Remove other than english

```python
import pandas as pd
from langdetect import detect, DetectorFactory
from langdetect.lang_detect_exception import LangDetectException

DetectorFactory.seed = 0

def is_english(text):
    try:
        return detect(text) == "en"
    except LangDetectException:  
        return False

df = reviews_df[reviews_df['content'].apply(is_english)]
```

```python
newdf = df.reset_index(drop=True)
```

```python
newdf
```

## Basic Sentiment Analysis

```python

def analyze_sentiment(review):
    blob = TextBlob(review)
    return blob.sentiment.polarity

reviews_df['sentiment_score'] = reviews_df['content'].apply(analyze_sentiment)

reviews_df['sentiment'] = reviews_df['sentiment_score'].apply(
    lambda x: 'Positive' if x > 0.2 else ('Negative' if x < -0.2 else 'Neutral')
)

sentiment_distribution = reviews_df['sentiment'].value_counts()
print("Sentiment Distribution:\n", sentiment_distribution)

positive_reviews = reviews_df[reviews_df['sentiment'] == 'Positive']['content']
negative_reviews = reviews_df[reviews_df['sentiment'] == 'Negative']['content']

def extract_topics(reviews, n_topics=3, n_words=5):
    vectorizer = CountVectorizer(stop_words='english', max_features=1000)
    dtm = vectorizer.fit_transform(reviews)
    lda = LatentDirichletAllocation(n_components=n_topics, random_state=42)
    lda.fit(dtm)
    topics = []
    for idx, topic in enumerate(lda.components_):
        topic_words = [vectorizer.get_feature_names_out()[i] for i in topic.argsort()[-n_words:]]
        topics.append(f"Topic {idx + 1}: {', '.join(topic_words)}")
    return topics

print("\nPositive Review Topics:")
positive_topics = extract_topics(positive_reviews, n_topics=3, n_words=5)
for topic in positive_topics:
    print(topic)

print("\nNegative Review Topics:")
negative_topics = extract_topics(negative_reviews, n_topics=3, n_words=5)
for topic in negative_topics:
    print(topic)

```

## Issue Labeling

```python
from transformers import pipeline, is_torch_available, is_tf_available

print("PyTorch available:", is_torch_available())
print("TensorFlow available:", is_tf_available())

classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")
```

```python
labels = ["app slowness", "booking problem", "offers issue", "customer service", "payment issue"]

reviews = df['content'].tolist()  
classified_reviews = []

for review in reviews:
    result = classifier(review, labels)
    classified_reviews.append({
        "review": review,
        "label": result['labels'][0],
        "score": result['scores'][0]
    })

classified_df = pd.DataFrame(classified_reviews)
```

```python
l = list(classified_df.columns)
l[2] = "Lable Confidence Score"
classified_df.columns = l
classified_df
```

```python
d = newdf[["at", "score","thumbsUpCount","reviewCreatedVersion", "replyContent", "repliedAt", "appVersion"]]
p = pd.concat([d,classified_df],axis=1)
```
